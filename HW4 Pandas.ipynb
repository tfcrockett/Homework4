```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Collection of information needed to create the summary of district

schools = school_data_complete['school_name'].unique()
total_schools = len(schools)
total_students = len(school_data_complete['student_name'])
school_budget=school_data_complete['budget'].unique()
total_budget = school_budget.sum()
average_math = school_data_complete['math_score'].mean()
average_reading = school_data_complete['reading_score'].mean()
passing_list_m = school_data_complete[school_data_complete.math_score>=70]
percent_math = len(passing_list_m)/total_students * 100
passing_list_r = school_data_complete[school_data_complete.reading_score>=70]
percent_reading = len(passing_list_r)/total_students * 100
overall_passing = (percent_math+percent_reading)/2
```

```python
#Dataframe of district: total schools,students, budget, average math and reading score, percentage passing and overall rate

district_summary = pd.DataFrame({"Total Schools":[total_schools],'Total Students':[total_students],'Total Budget':[total_budget],'Average Math Score':[average_math],
                    'Average Reading Score': [average_reading],'% Passing Math': [percent_math],'% Passing English': [percent_reading],
                     'Overall Passing Rage': [overall_passing]})
district_summary
```

```python
#counts the students in each school
school_count = school_data_complete['school_name'].value_counts()
schools_summary_table = pd.DataFrame(school_count)

#groups school by school name
grouped_school=school_data_complete.groupby('school_name')

schools_summary_table['type']= grouped_school['type'].unique()
#gets the budget of each school
schools_summary_table['budget']=grouped_school['budget'].unique()
schools_summary_table['student_budget']=schools_summary_table['budget']/schools_summary_table['school_name']

#gets the average scores for math and reading
schools_summary_table['average_math']=grouped_school['math_score'].mean()
schools_summary_table['average_reading']=grouped_school['reading_score'].mean()

#changes the column name
schools_summary_table.rename(columns = {'school_name':'Total Students'},inplace=True)

#gets the passing rate for math and reading
temp_math = school_data_complete.loc[school_data_complete['math_score']>=70]
pcent_math=temp_math.groupby('school_name')['math_score'].count()
total_students = schools_summary_table['Total Students']
schools_summary_table['percent_math']=pcent_math/total_students*100

temp_reading=school_data_complete.loc[school_data_complete['reading_score']>=70]
pcent_reading=temp_reading.groupby('school_name')['reading_score'].count()
schools_summary_table['percent_reading']=pcent_reading/total_students*100

#gets the overall passing rate
schools_summary_table['% Overall Passing Rate']=(schools_summary_table['percent_math']+schools_summary_table['percent_reading'])/2
schools_summary_table.head(5)
```
